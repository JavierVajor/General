### Rellena los NaN con 0 u otro valor SÓLO EN LA COLUMNA DESEADA

````python
data4=data
````

````python
data4['body']=data4['body'].fillna(0) #Con esto sobreescribimos los valores NaN de la columna deseada
data4['home.dest']=data4['home.dest'].fillna('Desconocido')
data4.head()
````

### Rellena los NaN con la media, mediana o moda de su columna

````python
media=data4['age'].mean() #mean,median,mode
````

````python
data4['age'].fillna(media) #Como se puede ver, el valor 15, que era NaN, toma el valor medio (29.88)
````

### Rellena los NaN con el mismo valor anterior (forward fill) o posterior (backward fill)

````python
data5=pd.read_csv(url)
data5['age'].fillna(method='ffill') #Como se puede ver, el valor 15, que era NaN, toma el valor anterior conocido (80)
````

````python
data6=pd.read_csv(url)
data6['age'].fillna(method='backfill') #Como se puede ver, el valor 15, que era NaN, toma el valor siguiente conocido (24)
````

## Variables Dummy

Una variable dummy permite eliminar una variable categórica, como puede ser el Sexo (H o M) y crear dos variables:
+ Male = 1 ó 0
+ Female = 0 ó 1

````python
def create_dummies(data, var_name):
    dummy=pd.get_dummies(data[var_name], prefix=var_name)
    data=data.drop(var_name, axis=1)
    data=pd.concat([data,dummy], axis=1)
    return data
````

````python
data7=data
````

````python
create_dummies(data7,'sex')
````
