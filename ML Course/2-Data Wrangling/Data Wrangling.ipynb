### Creación de un subconjunto de datos a partir del data frame

#### Así extraemos una columna

```python
account_lenght = data["Account Length"]
account_lenght.head()
```

#### Así extraemos un grupo de columnas

```python
desired_columns=[data.columns.values[1],"Phone", "Eve Charge", "Day Calls"]
subset=data[desired_columns]
subset.head()
```

#### Así extraemos el subset complemetario, es decir, el dataframe sin las columnas deseadas

En ocasiones, las columnas que queremos son más que las que no queremos. Por tanto, es más eficiente generar un subset complementario. Es decir, un subset que contenga todo, menos las de la lista que le damos

```python
all_columns=data.columns.values.tolist()
not_desired=["State","Area Code", "Phone",]
desired = list(set(all_columns) - set(not_desired))
subset=data[desired]
subset.head()
```

#### Crear subconjunto de filas y columnas al mismo tiempo

A traves de dos corchetes podemos indicar las columnas y las filas, respectivamente

```python
subset_3=data[["State", "Day Mins", "Night Mins"]] [:10]
subset_3
```

A traves de la función .iloc podemos acceder a las filas y columnas, respectivamente, proporcionando sus índices

```python
data.iloc[0:10, 3:7]  #10 primeras filas, columnas 3-7
```

A traves de la función .loc podemos acceder a las filas y columnas, respectivamente, proporcionando sus etiquetas

```python
data.loc[:5, ("Phone", "Day Mins")]
```

## Generador de números aleatorios

```python
import numpy as np
```

#### Generar un sólo número

```python
np.random.randint(1,100) #Genera un entero entre a y b
```

```python
np.random.random() # Genera un float entre 0 y 1
```

#### Generar una lista

```python
def randint_list (n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list (10, 1, 1000)
```

#### Generar múltiplos aleatorios de un número

```python
import random as rd
```

```python
rd.randrange(0,1000,13)  #(Multiplos de 13) + 0  en el rango de 0-1000
```

```python
rd.randrange(2,50,13)  #(Multiplos de 13) + 2  en el rango de 0-50
```

## Shuffilng o barajado

#### Para desordenar una lista

```python
lista = list(range(1,11))  #Crea una lista del 1 al 10
np.random.shuffle(lista)  #Desordena esa lista
lista
```

#### Para escoger un valor aleatoriamente de una lista

```python
rd.choice(lista)
```

## Seed o semilla: origen de la generación aleatoria

```python
np.random.seed(2020)   #Al determinar la semilla, siempre que ejecutemos numeros aleatorios, saldrán los mismos.
for i in range(3):
    print(np.random.random()) #Siempre se imprimirán los mismos numeros cada vez que ejecutemos
```
