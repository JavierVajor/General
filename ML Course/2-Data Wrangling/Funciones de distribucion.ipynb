## La simulación de Monte Carlo

```python
total_dardos = 5000
dardos_dentro = 0
coord_X = []
coord_Y = []
for dardo in range(total_dardos):
    coord_x_dardo = np.random.uniform(-1, 1)
    coord_X.append(coord_x_dardo)
    coord_y_dardo = np.random.uniform(-1, 1)
    coord_Y.append(coord_y_dardo)
    objetivo_dardo = coord_x_dardo**2 + coord_y_dardo**2
    if objetivo_dardo <= 1:
        dardos_dentro += 1
pi_aprox = dardos_dentro / total_dardos * 4
pi_aprox
```

```python
plt.rc('figure', figsize=(8,8))
plt.scatter(coord_X,coord_Y)
plt.show()
```

```python
coord_X_dentro, coord_Y_dentro = [],[] 
for i in range(total_dardos):
    if coord_X[i] * coord_X[i] + coord_Y[i] * coord_Y[i] <=1:
        coord_X_dentro.append(coord_X[i])
        coord_Y_dentro.append(coord_Y[i])
plt.scatter(coord_X,coord_Y)
plt.scatter(coord_X_dentro, coord_Y_dentro, color='r')
plt.show()
```
